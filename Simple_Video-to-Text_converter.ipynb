```python
import speech_recognition as sr 
import moviepy.editor as mp
```

### Now in the next step we'll be converting our video recording into an audio file from .mov format to .wav
### Now, we have some idea about both formats. It’s time to do the conversion using MoviePy library. You will not believe how easy it is.

```python
clip = mp.VideoFileClip('movie.mov') 
 
clip.audio.write_audiofile('converted.wav')
```

## Define SpeechRecogniser

```python
r = sr.Recognizer()
```

```python
audio = sr.AudioFile("converted.wav")
```

```python
with audio as source:
  audio_file = r.record(source)
result = r.recognize_google(audio_file)
```

```python
# exporting the result 
with open('recognized.txt',mode ='w') as file: 
   file.write("Recognized Speech:") 
   file.write("\n") 
   file.write(result) 
   print("ready!")
```
